1. O que é uma imagem digital?

Uma imagem digital é um conjunto de pixels, onde cada pixel é a menor unidade da imagem. Cada pixel contém informações sobre a cor e a intensidade da luz naquele ponto específico da imagem. O pixel é como se fosse o átomo da imagem!

</p>

Esses pixels são organizados em uma grade bidimensional, formando a imagem completa. A resolução de uma imagem é determinada pelo número de pixels na largura e na altura da imagem. Quanto maior o número de pixels, maior a resolução e, consequentemente, mais detalhada será a imagem. Você já ouviu falar de 720p, 1080p, ou 4K? Então, esses são nomes comerciais de diferentes resoluções de imagem, para 1280 x 720 pixels (720p), 1920 x 1080 (1080p) pixels e 3840 x 2160 pixels (4K).

Além disso, em vez de ver uma paisagem ou um retrato como nós vemos, o computador só processa números. Desse modo, uma imagem digital é na verdade uma matriz de vários números, o qual um computador é capaz de processar. O pixel, que é a menor unidade da imagem, também é um conjunto de números.

Um pixel é apenas um conjunto de informações (geralmente números) que representam as características de uma cor e sua intensidade, e, junto com milhares de outros pixels forma a imagem que vemos numa tela.

3. Escalas de cores e Valores de um Pixel:

- Escala Cinza (Grayscale):

  Nessa escala cada pixel tem um valor que vai de 0 (preto) a 255 (branco). Então, se olharmos para a tabela da imagem em grayscale, vamos ver números que vão de 0 a 255, indicando o quão claro ou escuro é aquele ponto. Esse tipo de escala é ideal para simplificar a análise de imagens. Também é comum normalizar os valores de cada pixel de 0 a 1.

- RGB (Red, Green, Blue):

  Na escala RGB, cada pixel é representado como um vetor com 3 posições, onde cada posição corresponde a uma das cores primárias: vermelho (R), verde (G) e azul (B). Cada uma dessas posições pode ter um valor que varia de 0 a 255, indicando a intensidade daquela cor. Por exemplo, um pixel com R=255, G=0, B=0 será vermelho puro, enquanto R=0, G=255, B=0 será verde puro. A combinação desses três valores define a cor exata de cada pixel. Esse é o modo de cor das imagens coloridas que estamos acostumados a ver em telas digitais.

Agora vamos visualizar melhor como o computador "entende" diferentes imagens. Vejamos a imagem abaixo em escala de cinza (grayscale).

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

image_url = "https://drive.google.com/uc?export=view&id=1tMOmy-nP0s3vHJwFhVgMvkl9jHTNSgOK"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content)).convert("L")  # Escala de cinza

# Converter para array
pixel_matrix = np.array(image)

# Exibir imagem
plt.figure(figsize=(6, 6))
plt.imshow(pixel_matrix, cmap="gray")
plt.colorbar(label="Intensidade do pixel")
plt.title("Imagem em escala de cinza")
plt.axis("off")
plt.show()

# Mostrar matriz
print("Matriz de pixels:")
print(pixel_matrix)
```

Resposta:

Agora é sua vez de investigar a matriz. Analise a matriz numpy da imagem para responder aos seguintes exercícios:

```python
# Exemplo resolvido: Qual é a intensidade média de pixel da primeira linha da imagem?
print(f"\n Exemplo.Intensidade média da primeira linha: {np.mean(pixel_matrix[0, :]):.2f}")

# 1. Quais são as dimensões da imagem em largura x altura?
altura, largura = pixel_matrix.shape # usar shape pra altura e largura da figura
print(f"\n 1.Dimensões da imagem: {largura} x {altura} (largura x altura)")

# 2. Imprima a primeira linha da matriz de pixels.
print(f"\n 2.Primeira linha da matriz: {pixel_matrix[0]}")

# 3. Qual a intensidade de pixel média da imagem?
media_pixels = np.mean(pixel_matrix) # usar mean pra intensidade média de pixels
print(f"\n 3.Intensidade de pixel média da imagem: {media_pixels:.2f}")

# 4. Qual é o valor do pixel mais claro e do mais escuro na imagem?
pixel_mais_claro = np.max(pixel_matrix) # usar .max pra pixel mais escuro
pixel_mais_escuro = np.min(pixel_matrix) # usar .min pra pixel mais claro

print(f"\n 4.Pixel mais claro (valor máximo): {pixel_mais_claro}")
print(f" 4.Pixel mais escuro (valor mínimo): {pixel_mais_escuro}")

# 5. Quantos pixels possuem intensidade acima da média?
pixels_acima_da_media = pixel_matrix > media_pixels
quantidade_pixels_acima_da_media = np.sum(pixels_acima_da_media)

print(f"\n 5.Número de pixels com intensidade acima da média: {quantidade_pixels_acima_da_media}")

# 6. A imagem contém mais áreas escuras ou claras? (Comparação de pixels abaixo e acima da média)
pixels_acima = pixel_matrix > media_pixels
pixels_abaixo = pixel_matrix < media_pixels

quantidade_acima = np.sum(pixels_acima)
quantidade_abaixo = np.sum(pixels_abaixo)

print(f"\n 6.Pixels acima da média (claros): {quantidade_acima}")
print(f" 6.Pixels abaixo da média (escuros): {quantidade_abaixo}")

mensagem_resultado = " A imagem tem quantidades iguais de áreas claras e escuras."

if quantidade_acima > quantidade_abaixo:
    mensagem_resultado = " A imagem contém mais áreas claras."
elif quantidade_acima < quantidade_abaixo:
    mensagem_resultado = " A imagem contém mais áreas escuras."

print(mensagem_resultado)

# 7. Plote a o trecho retangular da imagem entre os pixels 300 e 600 no eixo horizontal e 250 a 500 no eixo vertical.
print(f"\n 7.Imagem entre os pixels 300 e 600 no eixo horizontal e 250 a 500 no eixo vertical:")

recorte_imagem = pixel_matrix[250:500, 300:600]
plt.figure(figsize=(6, 6))
plt.imshow(recorte_imagem, cmap="gray")
plt.title("Trecho da imagem (x: 300–600, y: 250–500)")
plt.axis("off")
plt.colorbar(label="Intensidade do pixel")
plt.show()
```

Diferentemente da escala de cinza, em que cada pixel continha um único valor de intensidade (0 a 255), no RGB cada pixel será definido por 3 valores de intensidade: um para o vermelho (R), um para o verde (G) e um para o azul (B). Vamos ver a matriz de pixels dessa imagem.

```python
# Carrega o arquivo de imagem
image_url = "https://drive.google.com/uc?export=view&id=15soQXOGgPGZ5BCg8g6ZWIfIjUuHmsUgr"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content)).convert("RGB")

# Converte para a forma vetorial
pixel_matrix_rgb = np.array(image)

# Plota a imagem
plt.figure(figsize=(6, 6))
plt.imshow(pixel_matrix_rgb)
plt.title("Imagem RGB")
plt.show()

# Imprime a matriz de pixels
print("Matriz de pixels:")
print(pixel_matrix_rgb)
```

Note como os valores de pixel da primeira linha da matriz formam as cores que enxergamos na imagem:
<pre>  
[ [255   0   0]  -> 255 R, 0   G e 0   B -> 100% Vermelho
  [  0 255   0]  -> 0   R, 255 G e 0   B -> 100% Verde
  [  0   0 255]  -> 0   R, 0   G e 255 B -> 100% Azul
  [255 255   0]] -> 255 R, 255 G e 0   B -> 50%  Vermelho 50% Verde -> Amarelo (Composição aditiva)

Vamos ver como fica nossa imagem em escala de cinza carregada na escala RGB.

```python
# Carrega o arquivo de imagem
image_url = "https://drive.google.com/uc?export=view&id=1tMOmy-nP0s3vHJwFhVgMvkl9jHTNSgOK"
response = requests.get(image_url)
image = Image.open(BytesIO(response.content)).convert("RGB")

# Converte para a forma vetorial
pixel_matrix_rgb = np.array(image)

# Plota a imagem
plt.figure(figsize=(6, 6))
plt.imshow(pixel_matrix_rgb)
plt.title("Imagem RGB")
plt.show()
```

Agora é sua vez de trabalhar com imagens RGB. Usando a imagem acima, responda às questões:

```python
# Exemplo resolvido: Qual é o canal de cor dominante na imagem?
mean_channels = np.mean(pixel_matrix_rgb, axis=(0, 1))
dominant_channel = ["Red", "Green", "Blue"][np.argmax(mean_channels)]
print(f"\n Exemplo.Canal dominante: {dominant_channel}")

# 1. Qual é a média de intensidade de cada canal de cor (R, G, B)?
media_r, media_g, media_b = mean_channels

print(f"\n 1.Média do canal Vermelho (R): {media_r:.2f}")
print(f" 1.Média do canal Verde (G): {media_g:.2f}")
print(f" 1.Média do canal Azul (B): {media_b:.2f}")

# 2. Qual a proporção de pixels em que o vermelho é o canal mais intenso?
canal_vermelho = pixel_matrix_rgb[:, :, 0]
canal_verde = pixel_matrix_rgb[:, :, 1]
canal_azul = pixel_matrix_rgb[:, :, 2]

vermelho_dominante = (canal_vermelho > canal_verde) & (canal_vermelho > canal_azul)
pixels_vermelho_dominante = np.sum(vermelho_dominante)
pixels_na_imagem = pixel_matrix_rgb.shape[0] * pixel_matrix_rgb.shape[1]
proporcao_vermelho_dominante = pixels_vermelho_dominante / pixels_na_imagem

print(f"\n 2. Proporção de pixels em que o vermelho é o canal mais intenso: {proporcao_vermelho_dominante:.2f}")

# 3. Plote 3 imagens, cada uma mostrando apenas um canal de cor (R, G, B)
print(f"\n 3. Canal de cor (R, G, B) das imagens:")

imagem_vermelha = np.zeros_like(pixel_matrix_rgb)
imagem_vermelha[:, :, 0] = canal_vermelho  # mantém só o vermelho

imagem_verde = np.zeros_like(pixel_matrix_rgb)
imagem_verde[:, :, 1] = canal_verde  # mantém só o verde

imagem_azul = np.zeros_like(pixel_matrix_rgb)
imagem_azul[:, :, 2] = canal_azul  # mantém só o azul

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imagem_vermelha)
plt.title("Canal Vermelho (R)")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagem_verde)
plt.title("Canal Verde (G)")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagem_azul)
plt.title("Canal Azul (B)")
plt.axis("off")

plt.tight_layout()
plt.show()

# 4. Considerando que Vermelho e Verde formam tons quentes, e Azul os tons frios, a imagem tem mais tons frios ou quentes? (compare a média R+G vs B)
mean_r, mean_g, mean_b = mean_channels
media_quentes = mean_r + mean_g
media_frios = mean_b

mensagem_resultado = "A imagem tem quantidades equilibradas de tons quentes e frios."

if media_quentes > media_frios:
    mensagem_resultado = "A imagem tem predominância de tons quentes."
elif media_quentes < media_frios:
    mensagem_resultado = "A imagem tem predominância de tons frios."

print(f"\n 4.Tons frios ou quentes: {mensagem_resultado}")
print(f" 4.Média dos tons quentes (R + G): {media_quentes:.2f}")
print(f" 4.Média dos tons frios (B): {media_frios:.2f}")

# 5. Qual a variância da intensidade de cada canal de cor (R, G, B)? (pesquise sobre o conceito de variância)
print(f"\n 5.Variância da intensidade de cada canal de cor (R, G, B):")

variancia_canais = np.var(pixel_matrix_rgb, axis=(0, 1))

variancia_r, variancia_g, variancia_b = variancia_canais

print(f" 5.Variância do canal Vermelho (R): {variancia_r:.2f}")
print(f" 5.Variância do canal Verde (G): {variancia_g:.2f}")
print(f" 5.Variância do canal Azul (B): {variancia_b:.2f}")

# 6. Qual é a distribuição de intensidade dos canais RGB? (Histograma por canal)
print(f"\n 6.Distribuição de intensidade dos canais RGB:")

intensidade_vermelho = canal_vermelho.flatten()
intensidade_verde = canal_verde.flatten()
intensidade_azul = canal_azul.flatten()

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(intensidade_vermelho, bins=256, color='red', alpha=0.7)
plt.title("Histograma - Canal Vermelho (R)")
plt.xlabel("Intensidade")
plt.ylabel("Frequência")

plt.subplot(1, 3, 2)
plt.hist(intensidade_verde, bins=256, color='green', alpha=0.7)
plt.title("Histograma - Canal Verde (G)")
plt.xlabel("Intensidade")
plt.ylabel("Frequência")

plt.subplot(1, 3, 3)
plt.hist(intensidade_azul, bins=256, color='blue', alpha=0.7)
plt.title("Histograma - Canal Azul (B)")
plt.xlabel("Intensidade")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()

# 7. Qual a porcentagem de pixels acinzentados? (Pixels onde os valores de R, G e B são próximos - tente diferença menor que 10 para todo par de canais)
# pixel é considerado acinzentado se a diferença entre todos os pares de canais for menor que 10
acinzentados = (np.abs(canal_vermelho - canal_verde) < 10) & (np.abs(canal_vermelho - canal_azul) < 10) & (np.abs(canal_verde - canal_azul) < 10)
total_acinzentados = np.sum(acinzentados)
porcentagem_acinzentados = (total_acinzentados / pixels_na_imagem) * 100

print(f"\n 7.Porcentagem de pixels acinzentados: {porcentagem_acinzentados:.2f}")

# 8. Existe algum canal de cor saturado em alguma região da imagem? (Se há pixels com valores máximos em R, G ou B)
canais = [
    ("Vermelho (R)", canal_vermelho),
    ("Verde (G)", canal_verde),
    ("Azul (B)", canal_azul)
]

print(f"\n 8.Algum canal de cor saturado em alguma região da imagem:")

for nome_canal, matriz_canal in canais:
    if np.any(matriz_canal == 255):
        print(f" 8.Canal {nome_canal} possui pixels saturados (valor 255)")
    else:
        print(f" 8.Canal {nome_canal} não possui pixels saturados")
```

## Leitura opcional: outras escalas de cores
- CMYK (Cyan, Magenta, Yellow, Black):

  Este é o modo de cor utilizado em impressoras, onde as cores são criadas através da combinação de quatro valores: ciano (C), magenta (M), amarelo (Y) e preto (K). Cada uma dessas cores é representada por um valor que vai de 0 a 100, indicando a porcentagem de saturação dessa cor. Por exemplo, um valor de C=100, M=0, Y=0, K=0 indica um ciano puro, enquanto C=0, M=0, Y=0, K=100 resulta em preto puro. O modo CMYK permite reproduzir uma ampla gama de cores, incluindo diferentes tonalidades de cinza e preto, ajustando os valores de C, M, Y e K.

- Lab (Luminance, a, b):

Este é um modo de cor que se baseia na forma como o olho humano percebe as cores, dividindo a cor em três componentes: L para luminância (ou brilho), que varia de 0 a 100, e duas dimensões de cor oposta: a (que vai do verde ao vermelho) e b (que vai do azul ao amarelo), com valores que variam de -128 a 127 para cada componente. O modo Lab é usado principalmente em aplicações de gerenciamento de cores e é amplamente suportado por softwares profissionais de edição de imagem.

Qual a diferença entre luminância e brilho?

A luminância é uma medida objetiva da quantidade de luz que é refletida por uma superfície ou emitida por uma fonte de luz. O brilho, por outro lado, é uma percepção subjetiva da intensidade dessa luz, que pode variar dependendo do contexto visual e da sensibilidade de cada pessoa!

- HSV (Hue, Saturation, Value):

Este modo de cor é baseado nas propriedades perceptuais da cor e é amplamente utilizado em gráficos e computação gráfica. Ele define cores usando três dimensões: matiz (H), ou tonalidade, que varia de 0 a 360 graus e indica o tipo de cor (por exemplo, vermelho, verde, azul); saturação (S), que varia de 0% a 100% e indica a intensidade da cor; e valor (V), que também varia de 0% a 100% e representa o brilho da cor. Essa combinação de H, S e V permite definir cores de forma que seja intuitiva para o olho humano.

- YCbCr (Luminance, Chrominance Blue, Chrominance Red):

Este modo de cor é utilizado em sistemas de transmissão de vídeo, tanto analógico quanto digital. Ele separa as informações de cor em três componentes: Y para luminância (brilho), que varia de 0 a 255, Cb para crominância azul (diferença entre o azul e a luminância), e Cr para crominância vermelha (diferença entre o vermelho e a luminância), ambos variando de 16 a 240. O YCbCr é amplamente utilizado em dispositivos de vídeo, como DVD players, Blu-ray players e TVs, pois permite uma transmissão eficiente e compacta das informações de cor.
